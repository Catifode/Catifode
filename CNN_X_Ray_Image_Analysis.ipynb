```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense, Conv2D, MaxPool2D, Dropout, Flatten
from keras.models import Sequential
from keras.preprocessing import image
```

```python
train_datagen = image.ImageDataGenerator(
    rescale = 1/255, horizontal_flip = True , zoom_range = 0.2, shear_range = 0.2
)

train_data = train_datagen.flow_from_directory(directory="/content/drive/MyDrive/Radiology_Image/analysis/Train", target_size= (256,256), batch_size= 16,
                                               class_mode = 'binary')
```

```python
train_data.class_indices
```

```python
test_datagen = image.ImageDataGenerator(
    rescale = 1/255,
)
test_data = test_datagen.flow_from_directory(directory="/content/drive/MyDrive/Radiology_Image/analysis/Val", target_size= (256,256), batch_size= 16 ,
                                             class_mode = 'binary')
```

```python
model = Sequential()

model.add(Conv2D(filters= 32, kernel_size=(3,3), activation='relu', input_shape = (256,256,3)))

model.add(Conv2D(filters= 64, kernel_size=(3,3), activation='relu'))
model.add(MaxPool2D())
model.add(Dropout(rate=0.25))

model.add(Conv2D(filters= 64, kernel_size=(3,3), activation='relu'))
model.add(MaxPool2D())
model.add(Dropout(rate=0.25))

model.add(Conv2D(filters= 128, kernel_size=(3,3), activation='relu'))
model.add(MaxPool2D())
model.add(Dropout(rate=0.25))

model.add(Flatten())
model.add(Dense(units = 64, activation= 'relu'))
model.add(Dropout(rate=0.50))
model.add(Dense(units= 1, activation= 'sigmoid'))


model.compile(loss= keras.losses.binary_crossentropy, optimizer = "adam", metrics = ['acc'])
```

```python
model.summary()
```

```python
model.fit_generator(train_data, steps_per_epoch= 8, epochs= 10, validation_steps= 2,
                    validation_data= test_data)
```
